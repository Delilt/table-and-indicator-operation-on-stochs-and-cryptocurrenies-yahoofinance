```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
from matplotlib import style
```

```python
style.use("ggplot")

# Başlangıç ve bitiş tarihlerini belirleyin
start = dt.datetime(2000, 1, 1)
end = dt.datetime(2024, 10, 31)

# Tesla verilerini alıyoruz
df = yf.download('BTC-USD', start=start, end=end)
df.head()
df.to_csv("BTC.csv")
```

```python
dfBTC = pd.read_csv("BTC.csv",parse_dates=True,index_col=0)
dfBTC.head(10)
```

```python
dfBTC = dfBTC.apply(pd.to_numeric, errors='coerce')
```

```python
plot = dfBTC.plot(title="BTC Stocks")
plt.show()
```

```python
dfBTC['100ma'] = dfBTC['Adj Close'].rolling(window=100).mean()
```

```python
plot = dfBTC['100ma'].plot(title="100ma")
```

```python
import matplotlib.pyplot as plt

# Fiyat grafiği
plt.figure(figsize=(10, 8))  # Grafik boyutunu ayarla

ax1 = plt.subplot2grid((6, 1), (0, 0), rowspan=5, colspan=1)
ax2 = plt.subplot2grid((6, 1), (5, 0), rowspan=1, colspan=1)

ax1.plot(df.index, df['Adj Close'], label='Adj Close')
ax1.plot(df.index, df['100ma'], label='100ma')
ax2.bar(df.index, df['Volume'], alpha=0.3, label='Volume')

ax1.legend()
ax1.set_title('Bitcoin Fiyatı ve 100 Günlük Hareketli Ortalama')
ax1.set_ylabel('Fiyat')
ax2.set_title('Bitcoin Hacmi')
ax2.set_ylabel('Hacim')

plt.tight_layout()  # Grafiğin düzenini ayarla
plt.show()  # Grafiği göster

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import pandas as pd

# Jupyter Notebook içinde grafik göstermek için ayar
output_notebook()

# Örnek verileri kullanarak bir DataFrame oluşturma
data = {
    'date': pd.date_range(start='2021-01-01', periods=100),
    'Adj Close': pd.Series(range(100)) + 50,  # Basit bir artan değer serisi
    '100ma': pd.Series(range(100)) + 45,      # Basit bir hareketli ortalama
    'Volume': pd.Series(range(100)) * 10       # Basit bir hacim serisi
}
df = pd.DataFrame(data)

# Veriyi kaynak olarak kullanma
source = ColumnDataSource(data=dict(x=df['date'], y=df['Adj Close'], volume=df['Volume'], ma=df['100ma']))

# Grafik oluşturma
p = figure(title="Bitcoin Fiyatı ve Hacmi", x_axis_label='Tarih', y_axis_label='Fiyat', x_axis_type='datetime')

# Fiyat çizgisi
p.line('x', 'y', source=source, line_width=2, legend_label="Adj Close", color='blue')
p.line('x', 'ma', source=source, line_width=2, color='orange', legend_label="100ma")

# Hacim çubukları
p.vbar(x='x', top='volume', source=source, width=0.1, alpha=0.3, legend_label="Volume", color='green')

# Grafiği gösterme
show(p)
```
